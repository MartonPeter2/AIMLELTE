```python
import numpy as np
import matplotlib.pyplot as plt

# Define the proposed loss function
def custom_loss_function(P, A, M, epsilon, k):
    # Max(M, A)
    max_M_A = np.maximum(M, A)
    # Smooth transition function (sigmoid-like)
    sigma = 1 / (1 + np.exp(-k * (P - (M - epsilon))))
    # Loss function
    L = max_M_A - P + (M - A - (max_M_A - P)) * sigma
    return L

# Define parameters
A = 1  # Observed value
M = 2  # Competitor benchmark
epsilon = 0.2  # Small positive value for transition
k = 10  # Sharpness parameter
P = np.linspace(0, 3, 300)  # Prediction range

# Compute the custom loss
L_custom = custom_loss_function(P, A, M, epsilon, k)

# Plot the custom loss function
plt.figure(figsize=(10, 6))
plt.plot(P, L_custom, label=f"A={A}, M={M}, epsilon={epsilon}, k={k}")
plt.title("Custom Loss Function for profit making policy")
plt.xlabel("P (Prediction)")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()
plt.show()

# Define parameters
A = 2  # Observed value
M = 1  # Competitor benchmark
epsilon = 0.2  # Small positive value for transition
k = 10  # Sharpness parameter
P = np.linspace(0, 3, 300)  # Prediction range

# Compute the custom loss
L_custom = custom_loss_function(P, A, M, epsilon, k)

# Plot the custom loss function
plt.figure(figsize=(10, 6))
plt.plot(P, L_custom, label=f"A={A}, M={M}, epsilon={epsilon}, k={k}")
plt.title("Custom Loss Function for loss making policy")
plt.xlabel("P (Prediction)")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()
plt.show()

# Define parameters
A = 1  # Observed value
M = 1  # Competitor benchmark
epsilon = 0.2  # Small positive value for transition
k = 10  # Sharpness parameter
P = np.linspace(0, 3, 300)  # Prediction range

# Compute the custom loss
L_custom = custom_loss_function(P, A, M, epsilon, k)

# Plot the custom loss function
plt.figure(figsize=(10, 6))
plt.plot(P, L_custom, label=f"A={A}, M={M}, epsilon={epsilon}, k={k}")
plt.title("Custom Loss Function for properly priced policy")
plt.xlabel("P (Prediction)")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()
plt.show()
```
